```python
def  top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
top_k(items,k)
```

```python
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
```

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] += 1
    return d
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
count_by_category(pairs)
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return total - sum(nums)
nums = [3, 0, 1]
missing_number(nums)
```

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    if x not in seen:
        seen.add(x)
    return list(seen)
print(add_unique(1))
print(add_unique(2))
```

```python
stream = [1,2,3,4,5]
w = 3
```

```python
def moving_average(stream, w):
    if not stream or w <= 0:
        return []
    
    result = []
    window_sum = 0
    
    for i, num in enumerate(stream):
        window_sum += num
        
        if i < w - 1:
            result.append(window_sum / (i + 1))
        else:
            result.append(window_sum / w)
            if i >= w:
                window_sum -= stream[i - w]
    
    return result
stream=[1,2,3,4] 
w=3
print(moving_average(stream,w))  
```
