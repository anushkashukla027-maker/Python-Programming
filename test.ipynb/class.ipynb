```python
class Animal:
    def __init__(self, name, species,vaccination):
        self.name = name
        self.species = species
        self.vaccination = vaccination

    def show_info(self):
        print("Name:", self.name)
        print("Species:", self.species)
        print("vaccination:",self.vaccination)

    def sound(self):
        print("Animal makes a sound")


class Pet(Animal):
    def __init__(self, name, species, owner,vaccination):
        super().__init__(name, species,vaccination)   
        self.owner = owner

    def show_pet_info(self):
        self.show_info()
        print("Owner:", self.owner)

    def sound(self):
        print(self.name, "is a pet and makes a cute sound!")


p1 = Pet("Tommy", "Dog", "Anushka","rabbies")

p1.show_pet_info()
p1.sound()
```
